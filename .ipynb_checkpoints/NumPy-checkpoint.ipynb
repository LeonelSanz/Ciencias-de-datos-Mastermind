## Operaciones con arrays

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5, 6]
```

```python
array_1 = np.array(lista_1)
array_2 = np.array(lista_2)
```

```python
array_1 * array_2
```

```python
[ num1 * num2 for num1, num2 in zip(lista_1, lista_2)]
```

```python
%%timeit
[ num1 * num2 for num1, num2 in zip(lista_1, lista_2)]
```

```python
%%timeit
array_1 * array_2
```
